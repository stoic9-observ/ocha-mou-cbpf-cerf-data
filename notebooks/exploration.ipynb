# UNOCHA DATA Exploration

1. Need to import two data sets: "project summary" contains all projects within each CBPF and "contribution" includes countries and contributions. 
2. Understanding the various columns to have proper filters. 

```python
import pandas as pd

PROJECT_SUMMARY_URL = "https://cbpfapi.unocha.org/vo1/odata/ProjectSummary?$format=csv"
CONTRIBUTION_URL = "https://cbpfapi.unocha.org/vo1/odata/Contribution?$format=csv&includeTransfer=1"

projects = pd.read_csv(PROJECT_SUMMARY_URL, low_memory=False)
contrib = pd.read_csv(CONTRIBUTION_URL, low_memory=False)

projects.shape, contrib.shape
```

```python
contrib.columns.tolist()
```

# We now have the variables to filter by. 

1. 'DonorName' 
2. 'FiscalYear' 

```python
usg_2026 = contrib[
    (contrib["DonorName"] == "United States") &
    (contrib["FiscalYear"] == 2026)
]

usg_2026
```

```python
usg_2026[
    [
        "PooledFundName",
        "PledgeAmt",
        "PaidAmt",
        "PledgeDate",
        "PaidDate",
    ]
]
```
